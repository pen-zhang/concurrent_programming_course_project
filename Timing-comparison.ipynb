```python
import os
import numpy
import matplotlib.pyplot as pp
import pandas as pd
import seaborn
import subprocess

# combining tree same NUM

NUM = 1000
# java
if os.path.isfile('java-combining.txt'):
    os.remove('java-combining.txt')
for th in range(4,100,2):
    os.system(f'java -ea TestCombiningTree {th} {NUM}')

# c#
if os.path.isfile('c-combining.txt'):
    os.remove('c-combining.txt')
for th in range(4,100,2):
    os.system(f'TestCB.exe {th} {NUM}')


# go
if os.path.isfile('go-combining.txt'):
    os.remove('go-combining.txt')
for th in range(4,100,2):
    os.system(f'CombiningTree {th} {NUM}')
```

```python
javatime = pd.read_csv('java-combining.txt', names = ['language', 'threads', 'counting_number', 'time'])
ctime = pd.read_csv('c-combining.txt', names = ['language', 'threads', 'counting_number', 'time'])
gotime = pd.read_csv('go-combining.txt', names = ['language', 'threads', 'counting_number', 'time'])

all_time=[]
all_time.append(javatime)
all_time.append(ctime)
all_time.append(gotime)

all_times = pd.concat(all_time)
# all_times.tail()

time_index = all_times.set_index(['language'])

pp.figure(figsize = (18,8))

langs = ['java', 'c#', 'go']

for lang in langs:
    data = time_index.loc[lang]
    pp.plot(data['threads'], data['time'])
pp.legend(langs)
```

```python
import os
import numpy
import matplotlib.pyplot as pp
import pandas as pd
import seaborn
import subprocess

# counting network same tokens

tokens = 100000
# java
if os.path.isfile('java-counting.txt'):
    os.remove('java-counting.txt')
th = 4
while th<100:
    os.system(f'java -ea TestCounting {th} {tokens}')
    th*=2

# c#
if os.path.isfile('c-counting.txt'):
    os.remove('c-counting.txt')
th = 4
while th<100:
    os.system(f'TestCN.exe {th} {tokens}')
    th*=2


# go
if os.path.isfile('go-counting.txt'):
    os.remove('go-counting.txt')
th = 4
while th<100:
    os.system(f'CounterNetwork {th} {tokens}')
    th*=2
```
